```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
budget_table = school_data_complete[["school_name", "budget"]]
budget_table.head()
update_table = budget_table.drop_duplicates()
update_table.head()
Total_budget = update_table["budget"].sum()

Count_schools = len(school_data_complete["school_name"].unique())
Total_Students = len(school_data_complete["Student ID"].unique())
#school_types=school_data_complete.groupby(["school_name"]).last()["type"]
#budgetlist = school_data_complete["budget"].unique()
#Total_budget = budgetlist.sum()
#Total_budget
Avg_Math = school_data_complete["math_score"].mean()
Avg_Reading =school_data_complete["reading_score"].mean()
Overall_Rate = (Avg_Math+Avg_Reading)/2

students_Passmath = school_data_complete.loc[school_data_complete["math_score"]>=70, ["student_name", "math_score"]]
Number_studentsPassmath = len(students_Passmath)
Percent_PassMath = (Number_studentsPassmath/Total_Students)*100
Percent_PassMath
students_Passread = school_data_complete.loc[school_data_complete["reading_score"]>=70, ["student_name", "reading_score"]]
Number_studentsPassread = len(students_Passread)
Percent_PassReading = (Number_studentsPassread/Total_Students)*100
Percent_PassReading

District_Summary = pd.DataFrame({"Total Schools": Count_schools, "Total Students":Total_Students, "Total Budget": Total_budget,
                                "Average Math Score": Avg_Math, "Average Reading Score": Avg_Reading, 
                                "% Passing Math": [Percent_PassMath], "% Passing Reading": [Percent_PassReading], 
                                "% Overall Passing Rate": [Overall_Rate]})

District_Summary["Total Budget"] = District_Summary["Total Budget"].astype(float).map("${:,.2f}".format)
District_Summary["Total Students"] = District_Summary["Total Students"].map("{:,}".format)
District_Summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python

School_Summarydata = school_data_complete.groupby("school_name")
School_Type = School_Summarydata["type"].unique()
School_size = School_Summarydata["size"].unique()
Total_Students = School_Summarydata["Student ID"].count()
Total_Budget = School_Summarydata["budget"].unique()
Per_StudentBudget = (Total_Budget/Total_Students)
Avg_MathScore = School_Summarydata["math_score"].mean()
Avg_ReadScore = School_Summarydata["reading_score"].mean()

School_MathPassing = school_data_complete.loc[school_data_complete["math_score"]>=70,["school_name","Student ID", "math_Score"]]
StudentCount_PassMath = School_MathPassing.groupby("school_name")
Students_PassMath = StudentCount_PassMath["Student ID"].count()
Percent_PassMath = (Students_PassMath/Total_Students)*100

School_ReadPassing = school_data_complete.loc[school_data_complete["reading_score"]>=70,["school_name","Student ID", "reading_Score"]]
StudentCount_PassRead = School_ReadPassing.groupby("school_name")
Students_PassRead = StudentCount_PassRead["Student ID"].count()
Percent_PassRead = (Students_PassRead/Total_Students)*100
Overall_Rate = (Percent_PassMath + Percent_PassRead)/2

School_Summary = pd.DataFrame({"School Type": School_Type.str.get(0), "Total Students": Total_Students, "Total School Budget":Total_Budget, 
                              "Per Student Budget": Per_StudentBudget, "Average Math Score": Avg_MathScore, 
                              "Average Reading Score": Avg_ReadScore, "% Passing Math": Percent_PassMath, 
                              "% Passing Reading": Percent_PassRead, "% Overall Passing Rate": Overall_Rate})

School_Summary
School_SummaryUpdate = School_Summary.sort_values("% Overall Passing Rate", ascending=False)
School_SummaryUpdate["Total School Budget"] = School_Summary["Total School Budget"].astype(float).map("${:,.2f}".format)
School_SummaryUpdate["Total Students"] = School_Summary["Total Students"].map("{:,}".format)
School_SummaryUpdate["Per Student Budget"] = School_Summary["Per Student Budget"].astype(float).map("${:,.2f}".format)
School_SummaryUpdate.head()
```
