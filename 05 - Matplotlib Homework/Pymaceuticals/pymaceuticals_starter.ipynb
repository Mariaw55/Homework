# PYMACEUTICALS OBSERVATIONS

Throughout the 45-day drug trial, Capomulin proved to be the most effective drug in reducing average tumor volumes in mice.  The drug a tumor volume decrease of -19%, while the other drugs had increases of 46% and greater.  Addtionally, Capomulin had a lower average number of metastatic sites, and also had the best survival rate in mice by the 45th day. This resulted in over 80% survival rate for mice treated with Capomulin by the end of the trial, while Ketapril, Placebo and Infubinol had less than 50% survival rates.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data_df=pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_clinicaltrial_df = pd.merge(clinical_trial_df, mouse_data_df, on="Mouse ID", how = "left")


# Display the data table for preview
combined_clinicaltrial_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Combinedtrial_grouped = combined_clinicaltrial_df.groupby(["Drug", "Timepoint"])
Combinedtrial_grouped.mean().reset_index()

Tumor_Response_df = pd.DataFrame(Combinedtrial_grouped["Tumor Volume (mm3)"].mean().reset_index())
Tumor_Response_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
Combinedtrial_grouped = combined_clinicaltrial_df.groupby(["Drug", "Timepoint"])
Combinedtrial_grouped.sem().reset_index()
# Convert to DataFrame
Tumor_Response_sem = pd.DataFrame(Combinedtrial_grouped["Tumor Volume (mm3)"].sem().reset_index())
Tumor_Response_sem.head()
# Preview DataFrame

```

```python
Tumor_Response_semUpdate = Tumor_Response_sem.pivot(index="Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
Tumor_Response_semUpdate
Tumor_Response_semReduced = Tumor_Response_semUpdate.loc[:, ["Capomulin", "Infubinol","Ketapril", "Placebo"]]
capomulin_sem = Tumor_Response_semUpdate["Capomulin"]
infubinol_sem = Tumor_Response_semUpdate["Infubinol"]
ketapril_sem = Tumor_Response_semUpdate["Ketapril"]
placebo_sem = Tumor_Response_semUpdate["Placebo"]
```

```python
# Minor Data Munging to Re-Format the Data Frames

Tumor_Response_dfUpdate = Tumor_Response_df.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
Tumor_Response_dfUpdate
```

```python
# Generate the Plot (with Error Bars)
time = np.arange(0,50,5)
capomulin_data = Tumor_Response_dfUpdate["Capomulin"]
infubinol_data = Tumor_Response_dfUpdate["Infubinol"]
ketapril_data = Tumor_Response_dfUpdate["Ketapril"]
placebo_data = Tumor_Response_dfUpdate["Placebo"]

capomulin = plt.errorbar(time, capomulin_data, yerr=capomulin_sem, marker='o', color = 'red', label='Capomulin',linestyle=':')
infubinol = plt.errorbar(time, infubinol_data, yerr=infubinol_sem, marker='^', color = 'blue',label = "Infubinol",linestyle=':')
ketapril = plt.errorbar(time, ketapril_data,yerr=ketapril_sem, marker="s", color = 'green', label="Ketapril",linestyle=':')
placebo = plt.errorbar(time, placebo_data, yerr=placebo_sem, marker="d", color = 'black', label="Placebo",linestyle=':')
# Show the Figure
plt.grid()
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (in Days)")
plt.ylabel("Tumor Volume (3mm)")
plt.legend (loc="best")
plt.xlim(-3, 50)
plt.show()
plt.savefig("../Images/Tumor_Response_to_Treatment.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
combined_clinicaltrial_df
Combinedtrial_grouped = combined_clinicaltrial_df.groupby(["Drug", "Timepoint"])
Combinedtrial_grouped.count().reset_index()
# Convert to DataFrame
Survival_Rate = pd.DataFrame(Combinedtrial_grouped["Mouse ID"].count().reset_index())
Survival_Rate.head()
# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
Survival_RateUpdate = Survival_Rate.pivot_table(index="Timepoint", columns = "Drug",values="Mouse ID")
Survival_RateUpdate

PercentSurvival_Capomulin = ((Survival_RateUpdate["Capomulin"])/(Survival_RateUpdate["Capomulin"].max()))*100
PercentSurvival_Infubinol = ((Survival_RateUpdate["Infubinol"])/(Survival_RateUpdate["Infubinol"].max()))*100
PercentSurvival_Ketapril = ((Survival_RateUpdate["Ketapril"])/(Survival_RateUpdate["Ketapril"].max()))*100
PercentSurvival_Placebo = ((Survival_RateUpdate["Placebo"])/(Survival_RateUpdate["Placebo"].max()))*100
```

```python
# Generate the Plot (Accounting for percentages)
time = np.arange(0,50,5)
SurvivalRate_Capomulin = plt.plot(time, PercentSurvival_Capomulin, color ='red', marker='o', linestyle=":", label="Capomulin")
SurvivalRate_Infubinol = plt.plot(time, PercentSurvival_Infubinol, color ='blue', marker='^', linestyle=":", label="Infubinol")
SurvivalRate_Ketapril = plt.plot(time, PercentSurvival_Ketapril, color ='green', marker='s', linestyle=":", label="Ketapril")
SurvivalRate_Placebo = plt.plot(time, PercentSurvival_Placebo, color ='black', marker='d', linestyle=":", label="Placebo")
plt.grid()
plt.title("Survival Rates of Mice")
plt.xlabel("Time (in Days)")
plt.ylabel("Survival Rate %")
plt.legend (loc="best")
# Save the Figure
plt.savefig("../Images/Survival_Rates_of_Mice.png")
# Show the Figure
plt.show()
```
